# Investigative Data Visualisation


Investigative data visualization is a critical process in data analysis that enables us to explore, understand, and interpret datasets effectively. By visualizing relationships, identifying patterns, and performing operations like querying and standardizing data, we can uncover valuable insights that drive informed decision-making. Today, we will focus on three practical methods that form the foundation of investigative data visualization:

---

### 1. **Scatter Plots with Matplotlib**
Scatter plots are an excellent tool for visualizing the relationship between two variables. Using `Matplotlib.pyplot`, we can create plots that help us observe correlations, trends, or outliers in the data. With simple customizations like axis labels and colors, scatter plots can tell a compelling story.  

**Key Idea**:  
- Helps identify relationships (e.g., positive, negative, or no correlation) between two numerical variables.

**Example Code**:
```python
plt.scatter(x=Housing_df["Names_of_column1"], y=Housing_df["Names_of_column2"])
plt.xlabel("x_axis_label")  # Label for the X-axis
plt.ylabel("y_axis_label") # Label for the Y-axis
plt.show()
```

---

### 2. **Querying a DataFrame**
Querying is a powerful way to filter and investigate specific subsets of a dataset. By applying logical conditions to the data, we can extract meaningful information and analyze it further.  

**Key Idea**:  
- Allows targeted analysis by selecting rows that meet specific conditions.

**Example Code**:  
```python
# Retrieve rows where a column's value is greater than 500
filtered_df = Name_of_dataframe.query("Names_of_column > 500")

# Retrieve rows where multiple conditions are true
filtered_df = Name_of_dataframe.query("Names_of_column1 == 5 & Names_of_column2 > 700_000")
```

---

### 3. **Unique Values vs. Z-Scores**
Understanding the distribution of data is crucial for investigative visualization. Two essential concepts here are **unique values** and **z-scores**:

- **Unique Values**: Refers to the distinct, non-repeating values in a dataset or column. It helps in summarizing and understanding data variability.
- **Z-Scores**: Also known as standard scores, z-scores measure how far a particular value is from the mean in terms of standard deviations. They are vital for standardizing data and comparing different distributions.

**Example**:
```python
# Get unique values from a column
unique_values = Housing_df["Column_Name"].unique()

# Compute z-scores and retrieve unique z-score values
from scipy.stats import zscore
z_scores = zscore(Housing_df["Column_Name"]).unique()
```

**Key Idea**:  
- Unique values help identify categories or levels in data.
- Z-scores aid in standardizing and comparing data effectively.

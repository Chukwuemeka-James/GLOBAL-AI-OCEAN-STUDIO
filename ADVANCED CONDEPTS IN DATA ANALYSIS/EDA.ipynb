# Introduction to Exploratory Data Analysis (EDA) 

### **Definition**
Exploratory Data Analysis (EDA) is a critical process in data science used to understand the dataset's structure, detect anomalies, and identify patterns or relationships among variables before modeling. It involves summarizing a dataset using statistical and visualization techniques to uncover insights.

### **Why It Matters**
1. **Understanding Data:** EDA helps identify missing values, inconsistencies, or anomalies in the dataset.
2. **Hypothesis Generation:** It aids in formulating hypotheses about the data and determining modeling approaches.
3. **Improved Data Quality:** By identifying issues such as outliers or incorrect data entries, EDA ensures the dataset is clean and reliable.
4. **Guiding Feature Engineering:** It provides insights into which features might be most impactful in predictive modeling.
5. **Enhancing Decision-Making:** Visualizations and summaries help stakeholders understand the data easily.

---

## **The Data Analysis Process**

1. **Collecting Data:**
   - Data can come from various sources like databases, APIs, files (CSV, Excel), or web scraping.
   - Tools like Pythonâ€™s `requests` and `pandas` libraries, or cloud storage integrations, are used to collect data programmatically.

2. **Cleaning Data:**
   - **Handling Missing Values:** Techniques include filling missing values with mean/median or dropping rows/columns.
   - **Removing Duplicates:** Ensures there is no redundant information in the dataset.
   - **Correcting Data Types:** Ensures numerical data isn't stored as text, and vice versa.
   - Tools: `pandas` provides methods like `.dropna()`, `.fillna()`, and `.astype()` for cleaning data.

3. **Analyzing Data:**
   - Statistical summaries (mean, median, variance) help understand the data distribution.
   - Correlation and covariance reveal relationships between variables.
   - Tools: Python functions like `pandas.describe()` and libraries like `scipy` and `statsmodels`.

4. **Visualizing Data:**
   - Visualization makes it easier to identify trends, patterns, and anomalies.
   - Examples: Scatter plots for relationships, histograms for distributions, and boxplots for outliers.
   - Tools: `matplotlib`, `seaborn`, and `plotly`.

---

## **Overview of Popular Python Libraries for EDA**

1. **Pandas:**
   - Provides data structures like DataFrames to manipulate and analyze data.
   - Key methods: `.head()`, `.info()`, `.describe()`, `.groupby()`, and `.merge()`.

2. **Matplotlib:**
   - A foundational library for creating static, animated, and interactive visualizations in Python.
   - Common functions: `plt.plot()`, `plt.scatter()`, `plt.bar()`, and `plt.show()`.

3. **Seaborn:**
   - Built on `matplotlib`, Seaborn simplifies creating visually appealing statistical graphics.
   - Common plots: `sns.heatmap()`, `sns.boxplot()`, `sns.pairplot()`, and `sns.distplot()`.

4. **NumPy (Supportive Role):**
   - Provides numerical computing capabilities like arrays and mathematical functions.
   - Useful for data manipulation during EDA.

# Typical Exploration Question

### **1. What is the distribution of salaries in USD?**
#### **Objective**:
Understand the spread of salaries to determine if there are outliers or skewness in the dataset.

#### **Analysis**:
- Use a histogram and boxplot to visualize the distribution.
- Calculate key statistics like mean, median, and standard deviation.


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Statistics
print(data['salary_in_usd'].describe())
```

```python
# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data['salary_in_usd'], kde=True, bins=20, color='blue')
plt.title('Distribution of Salaries in USD')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 3))
sns.boxplot(x=data['salary_in_usd'], color='orange')
plt.title('Boxplot of Salaries in USD')
plt.xlabel('Salary in USD')
plt.show()
```

### **2. How does experience level affect salary?**
#### **Objective**:
Examine how different experience levels impact salary in USD.

#### **Analysis**:
- Group data by `experience_level` and calculate the average salary.
- Visualize using a bar chart.

```python
# Group by experience level
experience_salary = data.groupby('experience_level')['salary_in_usd'].mean().sort_values()
experience_salary
```

```python
# Bar chart
plt.figure(figsize=(8, 5))
experience_salary.plot(kind='bar', color='green')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.show()
```

### **3. What is the relationship between remote work (`remote_ratio`) and salary?**
#### **Objective**:
Analyze how working remotely impacts salaries.

#### **Analysis**:
- Compare average salaries across different `remote_ratio` categories.
- Visualize using a bar chart.

```python
# Group by remote ratio
remote_salary = data.groupby('remote_ratio')['salary_in_usd'].mean()
remote_salary
```

```python
# Bar chart
plt.figure(figsize=(8, 5))
remote_salary.plot(kind='bar', color='purple')
plt.title('Average Salary by Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.show()
```

### **4. How does company size affect salary?**
#### **Objective**:
Understand if the size of a company influences employee salaries.

#### **Analysis**:
- Group data by `company_size` and calculate the average salary.
- Visualize using a bar chart.

```python
# Group by company size
company_salary = data.groupby('company_size')['salary_in_usd'].mean()
company_salary
```

```python
# Bar chart
plt.figure(figsize=(8, 5))
company_salary.plot(kind='bar', color='coral')
plt.title('Average Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.show()
```

### **5. How does employment type influence salary distribution?**
#### **Objective**:
Analyze how different types of employment (e.g., full-time, part-time, contract, freelance) affect salary levels.

#### **Analysis**:
- Compare salary distributions across employment types using boxplots for a detailed view.

```python
# Boxplot of salary by employment type without palette
plt.figure(figsize=(10, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=data)
plt.title('Salary Distribution by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.show()
```

### **6. Which job titles are most likely to be fully remote?**
#### **Objective**:
Understand which roles are more likely to have a `remote_ratio` of 100% (fully remote work).

#### **Analysis**:
- Filter the dataset for `remote_ratio` = 100 and analyze the most frequent job titles.
- Visualize the findings using a bar chart.


```python
# Filter for fully remote roles
fully_remote_jobs = data[data['remote_ratio'] == 100]['job_title'].value_counts().head()
fully_remote_jobs
```

```python
# Bar chart for fully remote job titles
plt.figure(figsize=(10, 5))
fully_remote_jobs.plot(kind='bar', color='lightblue')
plt.title('Top 5 Fully Remote Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.show()
```
